```python
import pandas as pd
from datetime import date, time, datetime
```

```python
df= pd.read_csv(r"Registro_detallado_de_validaciones(4).csv", sep=",")
df.head()
```

```python
df_General= df[['Num Linea','Parada','Desc Parada','Fecha y hora validacione','Numero de la tarjeta','Sentido Resumen']]
df_General.head()
```

```python
df_General.groupby('Num Linea').mean()
```

```python
df_General_201L=df_General[df_General['Num Linea']==201]
df_General_202L=df_General[df_General['Num Linea']==202]
df_General_206L=df_General[df_General['Num Linea']==206]
df_General_209L=df_General[df_General['Num Linea']==209]
df_General_257L=df_General[df_General['Num Linea']==257]
```

```python
df_General_209L.head()
```

```python
df_General_201L['HORAN'] = df_General_201L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
df_General_202L['HORAN'] = df_General_202L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
df_General_206L['HORAN'] = df_General_206L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
df_General_209L['HORAN'] = df_General_209L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
df_General_257L['HORAN'] = df_General_257L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
```

```python
df_General_206L.head()
```

```python
df_General_201L['HORANN'] = df_General_201L['HORAN'].str.extract('([0-9]+)')
df_General_202L['HORANN'] = df_General_202L['HORAN'].str.extract('([0-9]+)')
df_General_206L['HORANN'] = df_General_206L['HORAN'].str.extract('([0-9]+)')
df_General_209L['HORANN'] = df_General_209L['HORAN'].str.extract('([0-9]+)')
df_General_257L['HORANN'] = df_General_257L['HORAN'].str.extract('([0-9]+)')
```

```python
df_General_209L.head()
```

```python
df_General_209L.groupby('HORANN').count()
```

```python
df_18pm_201L = df_General_201L[df_General_201L['HORANN'] == '17']
df_18pm_202L = df_General_202L[df_General_202L['HORANN'] == '17']
df_18pm_206L = df_General_206L[df_General_206L['HORANN'] == '17']
df_18pm_209L = df_General_209L[df_General_209L['HORANN'] == '17']
df_18pm_257L = df_General_257L[df_General_257L['HORANN'] == '17']
```

```python
df_18pm_257L.head()
```

```python
df_18pm_257L.shape
```

```python
Col_pivot_201L=pd.pivot_table(df_18pm_201L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
Col_pivot_202L=pd.pivot_table(df_18pm_202L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
Col_pivot_206L=pd.pivot_table(df_18pm_206L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
Col_pivot_209L=pd.pivot_table(df_18pm_209L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
Col_pivot_257L=pd.pivot_table(df_18pm_257L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
```

```python
Col_pivot_257L.head()
```

```python
df_18menos_201L = df_General_201L[(df_General_201L['HORANN'] < '17') | (df_General_201L['HORANN'] > '17')]
df_18menos_202L = df_General_202L[(df_General_202L['HORANN'] < '17') | (df_General_202L['HORANN'] > '17')]
df_18menos_206L = df_General_206L[(df_General_206L['HORANN'] < '17') | (df_General_206L['HORANN'] > '17')]
df_18menos_209L = df_General_209L[(df_General_209L['HORANN'] < '17') | (df_General_209L['HORANN'] > '17')]
df_18menos_257L = df_General_257L[(df_General_257L['HORANN'] < '17') | (df_General_257L['HORANN'] > '17')]
```

```python
df_18menos_257L.head()
```

```python
df_merge_201L = pd.merge(df_18pm_201L, df_18menos_201L, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_merge_202L = pd.merge(df_18pm_202L, df_18menos_202L, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_merge_206L = pd.merge(df_18pm_206L, df_18menos_206L, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_merge_209L = pd.merge(df_18pm_209L, df_18menos_209L, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_merge_257L = pd.merge(df_18pm_257L, df_18menos_257L, on=['Numero de la tarjeta'], how='inner', indicator=True)
```
