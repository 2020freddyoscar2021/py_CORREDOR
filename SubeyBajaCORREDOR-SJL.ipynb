```python
import pandas as pd
from datetime import date, time, datetime
```

```python
df= pd.read_csv(r"Registro_detallado_de_validaciones(4).csv", sep=",")
df.head()
```

```python
df_General= df[['Num Linea','Parada','Desc Parada','Fecha y hora validacione','Numero de la tarjeta','Sentido Resumen']]
df_General.head()
```

```python
df_General.groupby('Num Linea').count()
```

```python
df_General_404L=df_General[df_General['Num Linea']==404]
df_General_405L=df_General[df_General['Num Linea']==405]
df_General_409L=df_General[df_General['Num Linea']==409]
df_General_412L=df_General[df_General['Num Linea']==412]
```

```python
df_General_412L.head()
```

```python
df_General_404L['HORAN'] = df_General_404L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
df_General_405L['HORAN'] = df_General_405L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
df_General_409L['HORAN'] = df_General_409L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
df_General_412L['HORAN'] = df_General_412L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
```

```python
df_General_412L.head()
```

```python
df_General_404L['HORANN'] = df_General_404L['HORAN'].str.extract('([0-9]+)')
df_General_405L['HORANN'] = df_General_405L['HORAN'].str.extract('([0-9]+)')
df_General_409L['HORANN'] = df_General_409L['HORAN'].str.extract('([0-9]+)')
df_General_412L['HORANN'] = df_General_412L['HORAN'].str.extract('([0-9]+)')
```

```python
df_General_412L.head()
```

```python
df_7am_404L = df_General_404L[df_General_404L['HORANN'] == '08']
df_7am_405L = df_General_405L[df_General_405L['HORANN'] == '08']
df_7am_409L = df_General_409L[df_General_409L['HORANN'] == '08']
df_7am_412L = df_General_412L[df_General_412L['HORANN'] == '08']
```

```python
df_7am_412L.head()
```

```python
df_7am_412L.shape
```

```python
Col_pivot_404L=pd.pivot_table(df_7am_404L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
Col_pivot_405L=pd.pivot_table(df_7am_405L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
Col_pivot_409L=pd.pivot_table(df_7am_409L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
Col_pivot_412L=pd.pivot_table(df_7am_412L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
```

```python
Col_pivot_412L.head()
```

```python
df_8mas_404L = df_General_404L[df_General_404L['HORANN'] > '08']
df_8mas_405L = df_General_405L[df_General_405L['HORANN'] > '08']
df_8mas_409L = df_General_409L[df_General_409L['HORANN'] > '08']
df_8mas_412L = df_General_412L[df_General_412L['HORANN'] > '08']
```

```python
df_8mas_412L.head()
```

```python
df_merge_404L = pd.merge(df_7am_404L, df_8mas_404L, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_merge_405L = pd.merge(df_7am_405L, df_8mas_405L, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_merge_409L = pd.merge(df_7am_409L, df_8mas_409L, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_merge_412L = pd.merge(df_7am_412L, df_8mas_412L, on=['Numero de la tarjeta'], how='inner', indicator=True)
```

```python
df_merge_412L.head()
```

```python
df_merge_404L['LOGICO'] = ''
df_merge_404L.loc[df_merge_404L['Desc Parada_x'] == df_merge_404L['Desc Parada_y'], 'LOGICO'] = '0'
df_merge_404L.loc[df_merge_404L['Desc Parada_x'] != df_merge_404L['Desc Parada_y'], 'LOGICO'] = '1'
```

```python
df_merge_405L['LOGICO'] = ''
df_merge_405L.loc[df_merge_405L['Desc Parada_x'] == df_merge_405L['Desc Parada_y'], 'LOGICO'] = '0'
df_merge_405L.loc[df_merge_405L['Desc Parada_x'] != df_merge_405L['Desc Parada_y'], 'LOGICO'] = '1'
```

```python
df_merge_409L['LOGICO'] = ''
df_merge_409L.loc[df_merge_409L['Desc Parada_x'] == df_merge_409L['Desc Parada_y'], 'LOGICO'] = '0'
df_merge_409L.loc[df_merge_409L['Desc Parada_x'] != df_merge_409L['Desc Parada_y'], 'LOGICO'] = '1'
```

```python
df_merge_412L['LOGICO'] = ''
df_merge_412L.loc[df_merge_412L['Desc Parada_x'] == df_merge_412L['Desc Parada_y'], 'LOGICO'] = '0'
df_merge_412L.loc[df_merge_412L['Desc Parada_x'] != df_merge_412L['Desc Parada_y'], 'LOGICO'] = '1'
```
