```python
import pandas as pd
from datetime import date, time, datetime
```

```python
df= pd.read_csv(r"Registro_detallado_de_validaciones(4).csv", sep=",")
df.head()
```

```python
df_General= df[['Num Linea','Parada','Desc Parada','Fecha y hora validacione','Numero de la tarjeta','Sentido Resumen']]
df_General.head()
```

```python
df_General.groupby('Num Linea').count()
```

```python
df_General_301L=df_General[df_General['Num Linea']==301]
df_General_302L=df_General[df_General['Num Linea']==302]
df_General_303L=df_General[df_General['Num Linea']==303]
df_General_306L=df_General[df_General['Num Linea']==306]
```

```python
df_General_301L.head()
```

```python
df_General_301L['HORAN'] = df_General_301L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
df_General_302L['HORAN'] = df_General_302L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
df_General_303L['HORAN'] = df_General_303L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
df_General_306L['HORAN'] = df_General_306L['Fecha y hora validacione'].str.extract(r'(\d\d:\d\d:\d\d)')
```

```python
df_General_306L.head()
```

```python
df_General_301L['HORANN'] = df_General_301L['HORAN'].str.extract('([0-9]+)')
df_General_302L['HORANN'] = df_General_302L['HORAN'].str.extract('([0-9]+)')
df_General_303L['HORANN'] = df_General_303L['HORAN'].str.extract('([0-9]+)')
df_General_306L['HORANN'] = df_General_306L['HORAN'].str.extract('([0-9]+)')
```

```python
df_General_306L.head()
```

```python
df_7am_301L = df_General_301L[df_General_301L['HORANN'] == '08']
df_7am_302L = df_General_302L[df_General_302L['HORANN'] == '08']
df_7am_303L = df_General_303L[df_General_303L['HORANN'] == '08']
df_7am_306L = df_General_306L[df_General_306L['HORANN'] == '08']
```

```python
df_7am_306L.head()
```

```python
df_7am_306L.shape
```

```python
Col_pivot_301L=pd.pivot_table(df_7am_301L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
Col_pivot_302L=pd.pivot_table(df_7am_302L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
Col_pivot_303L=pd.pivot_table(df_7am_303L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
Col_pivot_306L=pd.pivot_table(df_7am_306L,index=['Sentido Resumen', 'Desc Parada','Parada','Num Linea'], aggfunc='count')
```

```python
Col_pivot_306L.head()
```

```python
df_8mas_301L = df_General_301L[df_General_301L['HORANN'] > '08']
df_8mas_302L = df_General_302L[df_General_302L['HORANN'] > '08']
df_8mas_303L = df_General_303L[df_General_303L['HORANN'] > '08']
df_8mas_306L = df_General_306L[df_General_306L['HORANN'] > '08']
```

```python
df_8mas_306L.head()
```

```python
df_merge_301L = pd.merge(df_7am_301L, df_8mas_301L, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_merge_302L = pd.merge(df_7am_302L, df_8mas_302L, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_merge_303L = pd.merge(df_7am_303L, df_8mas_303L, on=['Numero de la tarjeta'], how='inner', indicator=True)
df_merge_306L = pd.merge(df_7am_306L, df_8mas_306L, on=['Numero de la tarjeta'], how='inner', indicator=True)
```

```python
df_merge_306L.head()
```

```python
df_merge_301L['LOGICO'] = ''
df_merge_301L.loc[df_merge_301L['Desc Parada_x'] == df_merge_301L['Desc Parada_y'], 'LOGICO'] = '0'
df_merge_301L.loc[df_merge_301L['Desc Parada_x'] != df_merge_301L['Desc Parada_y'], 'LOGICO'] = '1'
```

```python
df_merge_302L['LOGICO'] = ''
df_merge_302L.loc[df_merge_302L['Desc Parada_x'] == df_merge_302L['Desc Parada_y'], 'LOGICO'] = '0'
df_merge_302L.loc[df_merge_302L['Desc Parada_x'] != df_merge_302L['Desc Parada_y'], 'LOGICO'] = '1'
```

```python
df_merge_303L['LOGICO'] = ''
df_merge_303L.loc[df_merge_303L['Desc Parada_x'] == df_merge_303L['Desc Parada_y'], 'LOGICO'] = '0'
df_merge_303L.loc[df_merge_303L['Desc Parada_x'] != df_merge_303L['Desc Parada_y'], 'LOGICO'] = '1'
```

```python
df_merge_306L['LOGICO'] = ''
df_merge_306L.loc[df_merge_306L['Desc Parada_x'] == df_merge_306L['Desc Parada_y'], 'LOGICO'] = '0'
df_merge_306L.loc[df_merge_306L['Desc Parada_x'] != df_merge_306L['Desc Parada_y'], 'LOGICO'] = '1'
```
